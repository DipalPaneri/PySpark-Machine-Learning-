So for example, let's say we wanted to know the average closing price per year. Easy! With a groupby and the year() function call:

```python
df.withColumn("Year", year(df['Date'])).show()
```

```python
new_df=df.withColumn("Year", year(df['Date']))
```

```python
new_df.groupBy('Year').mean().select(['Year', "avg(Close)"]).show()
```

```python
result = new_df.groupBy('Year').mean().select(['Year', "avg(Close)"])
```

```python
result = result.withColumnRenamed("avg(Close)", "Avg Closing Price")
```

```python
result.select(['Year',format_number('Avg Closing Price', 2).alias("Avg Closing Price")]).show()
```
